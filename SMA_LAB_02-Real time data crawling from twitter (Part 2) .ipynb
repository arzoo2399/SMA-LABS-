### 5.Extract text, screen names and hashtags from tweets

```python
status_texts=[]
screen_names=[]
hashtags=[]
words=[]

for i in statuses:
    status_texts.append(i['text'])
    for j in i['entities']['user_mentions']:
        screen_names.append(j['screen_name'])
    for k in i['entities']['hashtags']:
        hashtags.append(k['text'])

for u in status_texts:
    for v in u.split():
        words.append(v)


# Explore the first 5 items for each...

print(json.dumps(status_texts[0:5], indent=1))
print(json.dumps(screen_names[0:5], indent=1) )
print(json.dumps(hashtags[0:5], indent=1))
print(json.dumps(words[0:5], indent=1))
```

### 6.Create a frequency distribution from the words of the tweets

```python
for item in [words, screen_names, hashtags]:
    c = Counter(item)
    print(c.most_common(10)) # top 10
    print()
```

### 7.Using prettytable library, print the above frequency distribution in the form of a table

```python
from prettytable import PrettyTable

for label, data in (('Word', words),('Screen Name', screen_names),('Hashtag', hashtags)):
    pt = PrettyTable(field_names=[label,'Count']) 
    c = Counter(data)
    for kv in c.most_common(10):
      pt.add_row(kv)
    pt.align[label], pt.align['Count'] = 'l', 'r' # Set column alignment
    print(pt)
```
